```python
def label_title(title):
    if title in ['Cực kì hài lòng', 'Hài lòng']:
        return 'Tốt'
    elif title == 'Bình thường':
        return 'Bình thường'
    else:
        return 'Xấu'
df['label'] =df['title'].apply(label_title)
# Hàm tùy chỉnh để gắn nhãn spam
def label_spam(content):
    if pd.isnull(content) or content.strip() == '':  # Kiểm tra nếu content là NaN hoặc rỗng
        return 1  # Spam
    else:
        return 0  # Không phải spam
# Thêm cột 'spam' vào DataFrame
df['spam'] = df['content'].apply(label_spam)


display(df)
```

```python
def preprocess_and_predict(new_df, stop_words, vectorizer, scaler, model_rf_spam):
    """
    Hàm xử lý văn bản, chuẩn hóa dữ liệu, kết hợp TF-IDF và rating,
    và dự đoán kết quả SPAM.

    Parameters:
    - new_df (pd.DataFrame): DataFrame chứa dữ liệu với các cột 'comment' và 'rating'.
    - stop_words (set): Tập các từ dừng (stop words).
    - vectorizer (TfidfVectorizer): Vectorizer đã được huấn luyện.
    - scaler (StandardScaler): Scaler đã được huấn luyện.
    - model_rf_spam (trained model): Mô hình đã được huấn luyện để dự đoán SPAM.

    Returns:
    - pd.DataFrame: DataFrame với cột 'predicted_spam' được thêm vào.
    """
    # Tách từ, chuyển chữ thường, loại bỏ stop words, và ký tự đặc biệt
    new_df['processed_comment'] = new_df['comment'].apply(lambda x: word_tokenize(x, format="text").lower())
    new_df['processed_comment'] = new_df['processed_comment'].apply(
        lambda x: ' '.join([word for word in x.split() if word not in stop_words])
    )
    new_df['processed_comment'] = new_df['processed_comment'].apply(
        lambda x: re.sub(r'[^a-zA-Zàáạảãâầấậẩẫăằắặẳẵèéẹẻẽêềếệểễìíịỉĩòóọỏõôồốộổỗơờớợởỡùúụủũưừứựửữỳýỵỷỹđ ]', '', x)
    )

    # Biểu diễn văn bản bằng TF-IDF (sử dụng vectorizer đã huấn luyện)
    X_text_new = vectorizer.transform(new_df['processed_comment'])

    # Chuẩn hóa cột rating
    X_rating_new = scaler.transform(new_df[['rating']])

    # Kết hợp TF-IDF và rating
    X_combined_new = sp.hstack([X_text_new, sp.csr_matrix(X_rating_new)])

    # Dự đoán SPAM
    y_pred_spam = model_rf_spam.predict(X_combined_new)

    # Thêm kết quả dự đoán vào DataFrame
    new_df['predicted_spam'] = y_pred_spam

    return new_df[['comment', 'rating', 'predicted_spam']]

```
